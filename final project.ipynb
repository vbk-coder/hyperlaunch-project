## Data cleaning

### Subtask:
Clean the cricket dataset by handling missing values and inconsistencies.


**Reasoning**:
Impute missing numerical values using the median, impute missing 'Year' values using the mode, and handle categorical features based on the number of unique values. Then, handle duplicates and verify data types.



```python
# Impute numerical features (excluding 'Year')
numerical_cols = df.select_dtypes(include=['number']).columns
for col in numerical_cols:
    if col != 'Year':
        df[col] = pd.to_numeric(df[col], errors='coerce')
        df[col] = df[col].fillna(df[col].median())

# Impute 'Year' column with the mode
df['Year'] = df['Year'].fillna(df['Year'].mode()[0])

# Handle categorical features
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    if df[col].nunique() < 10:  # Adjust threshold as needed
        df[col] = df[col].replace('No stats', df[col].mode()[0])
    else:
        df = df[df[col] != 'No stats']

# Handle duplicate rows
df.drop_duplicates(inplace=True)

# Verify data types
for col in numerical_cols:
    if col != 'Year':
        df[col] = pd.to_numeric(df[col])
    else:
        df[col] = pd.to_numeric(df[col], downcast='integer')

for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].astype('category')

# Explore cleaned data
print(df.info())
print(df.isnull().sum())

numerical_cols = df.select_dtypes(include=['number']).columns
for col in numerical_cols:
    if col != 'Year':
        plt.figure(figsize=(8, 6))
        sns.histplot(df[col], kde=True)
        plt.title(f"Distribution of {col} after cleaning")
        plt.show()
```
